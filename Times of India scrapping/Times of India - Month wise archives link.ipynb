```python
#importing libraries
import requests
from bs4 import BeautifulSoup
```

```python
#requesting to download the html content of the page
page = requests.get("https://timesofindia.indiatimes.com/archive.cms")
```

```python
#making soup object
soup = BeautifulSoup(page.content,'html.parser')
```

```python
#Getting tags which contain links of month wise archive page
link_tags = soup.find_all('a',class_ = 'normtxt')
```

```python
# extracting links from tags
month_wise_archive_link = []
for tag in link_tags:
    month_wise_archive_link.append(tag.get('href'))
```

```python
#bringing the links in actual format
for i in range(len(month_wise_archive_link)):
    month_wise_archive_link[i] = "https://timesofindia.indiatimes.com"+month_wise_archive_link[i]
```
